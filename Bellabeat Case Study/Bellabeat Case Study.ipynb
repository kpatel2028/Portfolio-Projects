We're seeing that all the sleep records has been recorded at **12:00:00 AM** only once per day. <br>
So, we're going to remove the time.

```python
#Function to select the date from the SleepDay column
def new_activity(row):
    row.SleepDay = row.SleepDay[:9]
    return row
sleep_day=sleep_day.apply(new_activity, axis="columns").rename(columns={"SleepDay":"ActivityDate"})
```

**Does total steps have any relation with sleep?**<br>
First, we have to make sure that there is no null value and keep only two column TotalMinutesAsleep & TotalSteps from dailyactivitywithsleep

```python
left = daily_activity.set_index(['Id','ActivityDate'])
right = sleep_day.set_index(['Id','ActivityDate'])
#combining daily_activity table with sleep_day table to extract TotalStep & TotalMinutesAsleep column.
dailyactivitywithsleep = left.join(right)
dailyactivitywithsleep.reset_index()
dailyactivitywithsleep=dailyactivitywithsleep.drop_duplicates(keep='first',inplace=False).reset_index()
stepsnsleep = dailyactivitywithsleep[['TotalSteps','TotalMinutesAsleep']].dropna(subset=['TotalMinutesAsleep'])
stepsnsleep.to_csv("stepsnsleep.csv") #downloading for spreadsheet program.
```

```python
stepsnsleep.head(300)
```

```python
#Plotting the graph.
ptl.figure(figsize=(18,6))
ptl.title("Total Minutes Asleep vs Total Steps")
sns.lineplot(x = dailyactivitywithsleep['TotalSteps'], y=dailyactivitywithsleep['TotalMinutesAsleep'])
blue, = sns.color_palette("muted", 1)
ptl.show()
```

Here we can see that people who sleeps more time approximately 800 minutes per day, their total steps is between 0-5000.<br> On the other hand, who steps daily 15000-20000 daily, their sleep time is minimum.

**Who walks more?**<br>
We have to analyze that women having different BMI scales, how much they walk?

```python
weight_log.head()
```

```python
def new_activity(row):
    row.Date = row.Date[:9]
    return row

#Joining weight_log table with dailyactivitywithsleep table to get the Best BMI & High BMI.

weight_log=weight_log.apply(new_activity, axis="columns").rename(columns={"Date":"ActivityDate"})
weight_log=weight_log[['Id','ActivityDate','BMI']]
left = weight_log.set_index(['Id','ActivityDate'])
right = dailyactivitywithsleep.set_index(['Id','ActivityDate'])
n=left.join(right)
stepswithBMI=n.reset_index()[["Id","ActivityDate","BMI","TotalDistance"]].dropna(subset=['TotalDistance'])
meanforbestBMI=stepswithBMI.loc[(stepswithBMI.BMI >= 18.5) & (stepswithBMI.BMI <= 24.9)].TotalDistance.mean()
meanforhighBMI=stepswithBMI.loc[(stepswithBMI.BMI > 24.9) & (stepswithBMI.BMI <= 29.9)].TotalDistance.mean()
output=pd.DataFrame({'Best BMI(18.5-24.9)': [meanforbestBMI], 'High BMI(24.9>)': [meanforhighBMI]})
output.to_csv("average_distance_with_BMI.csv")
```

```python
print(output)
```

```python
#Plotting the distance vs BMI Graph from the above output.

ptl.figure(figsize=(12,6))
ptl.title("Distance vs BMI")
##df = pd.DataFrame({'BMI': ['18.5 - 24.9', '24.9>']})
sns.barplot(data = output)
ptl.ylabel("Distance")
ptl.xlabel("BMI")
ptl.show()
```

From the graph, we can see that women with Best BMI range walks the distance of 6.71 km whereas High BMI range walks the distance of 13.11 km.
